```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, beta
```

```python
def compare(a1, b1, a2, b2):
    x = np.linspace(0,1,100)  
    beta1 = beta(a1, b1)
    beta2 = beta(a2, b2)

    print("beta1 {} {} ".format(beta1.mean(), beta1.std()))
    print("beta2 {} {} ".format(beta2.mean(), beta2.std()))
    
    plt.plot(x, beta1.pdf(x), color='orange')
    plt.plot(x, beta2.pdf(x), color='green')
    plt.show()
    
    n = 200000
    samples1 = beta1.rvs(size=n)
    samples2 = beta2.rvs(size=n)
    ptb = sum(np.greater_equal(samples1, samples2)) / n
    
    print("prob 1 beats 2 " + str(ptb))
```

```python
# sanity check
# about 50 ptb
compare(100, 900, 100, 900)
```

```python
# sanity check
# 1 < 2
compare(100, 900, 100 + 10, 900 - 10)
```

```python
# sanity check
# 1 > 2
compare(100 + 10, 900 - 10, 100, 900)
```

```python
# same mean
# variance 1 > 2
compare(100, 900, 1000, 9000)
```

```python
# now just add 1 extra conversion
# ptb jumps to almost 0.53
compare(100 + 1, 900 - 1, 1000, 9000)
```

```python
# add just 2 extra conversion
# ptb jumps almost 0.57
compare(100 + 2, 900 - 2, 1000, 9000)
```
